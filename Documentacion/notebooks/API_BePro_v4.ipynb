```python
def convertStringToFecha(start_time):
  try:
    fecha = start_time
    fecha = fecha.split(" ")[0]
    fecha = fecha.split("-")[2] + '/' + fecha.split("-")[1] + '/' + fecha.split("-")[0]
  except:
    fecha=''
  return fecha

def convertStringToHora(start_time):
  try:
    hora = start_time
    hora = hora.split(" ")[1]
    hora = hora.split(":")[0] + ':' + hora.split(":")[1] + ' hs'
  except:
    hora=''
  return hora
```
