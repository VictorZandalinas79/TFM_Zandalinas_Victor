```python
# 1. Primero, veamos qué tipo de datos tenemos en cada columna
print("Tipos de datos antes de la conversión:")
print(df_eventos_datos_acumulados.dtypes)

# 2. Función para convertir de manera segura los valores binarios
def convert_binary_string(x):
    try:
        if pd.isna(x):
            return 0
        if isinstance(x, (int, float)):
            return int(x)
        # Si es una cadena de 0s y 1s
        if isinstance(x, str) and all(c in '01' for c in x):
            return sum(int(d) for d in x)
        return 0
    except:
        return 0

# 3. Identificar columnas numéricas y no numéricas
columnas_no_numericas = ['match_id', 'team_id', 'equipo', 'jugador', 'demarcacion', 'liga', 'partido', 'jornada']
columnas_a_convertir = [col for col in df_eventos_datos_acumulados.columns if col not in columnas_no_numericas]

# 4. Convertir las columnas
for col in columnas_a_convertir:
    if col in df_eventos_datos_acumulados.columns:
        print(f"Convirtiendo columna: {col}")
        print(f"Valores únicos antes de la conversión: {df_eventos_datos_acumulados[col].unique()[:5]}")
        df_eventos_datos_acumulados[col] = df_eventos_datos_acumulados[col].apply(convert_binary_string)
        print(f"Valores únicos después de la conversión: {df_eventos_datos_acumulados[col].unique()[:5]}\n")

# 5. Convertir team_id a entero
df_eventos_datos_acumulados['team_id'] = df_eventos_datos_acumulados['team_id'].astype(float).astype(int)
df_team_stats_league_all['team_id'] = df_team_stats_league_all['team_id'].astype(int)

# 6. Agrupar y sumar los datos ya convertidos
df_eventos_agregados = df_eventos_datos_acumulados.groupby(['match_id', 'team_id'])[columnas_a_convertir].sum().reset_index()

# 7. Obtener las columnas no numéricas
if columnas_info := [col for col in columnas_no_numericas if col in df_eventos_datos_acumulados.columns and col not in ['match_id', 'team_id']]:
    df_eventos_info = df_eventos_datos_acumulados.groupby(['match_id', 'team_id'])[columnas_info].first().reset_index()
    df_eventos_agregados = pd.merge(df_eventos_agregados, df_eventos_info, on=['match_id', 'team_id'])

# 8. Eliminar columnas duplicadas y hacer el merge
columnas_duplicadas = list(set(df_eventos_agregados.columns) & set(df_team_stats_league_all.columns) - {'match_id', 'team_id'})
df_eventos_agregados_final = df_eventos_agregados.drop(columns=columnas_duplicadas)

df_equipos_estadisticas = pd.merge(
    df_eventos_agregados_final,
    df_team_stats_league_all,
    on=['match_id', 'team_id'],
    how='inner'
)

# 9. Verificar resultados
print("\nDimensiones del nuevo DataFrame:", df_equipos_estadisticas.shape)
print("\nMuestra de algunas columnas numéricas:")
columnas_muestra = [col for col in df_equipos_estadisticas.columns if col not in columnas_no_numericas][:5]
print(df_equipos_estadisticas[columnas_muestra].describe())
```

```python
# Definir los componentes de cada KPI
kpi_components = {
   "Posesion_Dominante": {
       "columnas": ["posesion", "pases_medios_exitosos", "pases_cortos_exitosos"],
       "pesos": [0.4, 0.3, 0.3]
   },
   "Estilo_Combinativo_Directo": {
       "columnas": ["pases_largos_exitosos", "duelos_aereos_ganados_zona_alta", "duelos_aereos_perdidos_zona_alta"],
       "pesos": [0.4, 0.4, -0.2]
   },
   "Estilo_Presionante": {
       "columnas": ["recuperaciones_zona_alta", "entradas_ganadas_zona_alta", "duelos_suelo_ganados_zona_alta"],
       "pesos": [0.4, 0.3, 0.3]
   },
   "Altura_Bloque_Defensivo": {
       "columnas": ["recuperaciones_zona_alta", "recuperaciones_zona_media", "recuperaciones_zona_baja"],
       "pesos": [0.5, 0.3, 0.2]
   },
   "Progresion_Ataque": {
       "columnas": ["pases_ultimo_tercio_exitosos", "pases_progresivos_finalizacion", "conducciones_finalizacion"],
       "pesos": [0.4, 0.3, 0.3]
   },
   "Verticalidad": {
       "columnas": ["pases_adelante_exitosos", "pases_progresivos_finalizacion"],
       "pesos": [0.5, 0.5]
   },
   "Ataques_Bandas": {
       "columnas": ["centros_exitosos", "regates_zona_creacion", "conducciones_carril_derecho", "conducciones_carril_izquierdo"],
       "pesos": [0.3, 0.3, 0.2, 0.2]
   },
   "Rendimiento_Finalizacion": {
       "columnas": ["goles", "disparos_a_puerta", "xg"],
       "pesos": [0.4, 0.3, 0.3]
   },
   "Peligro_Generado": {
       "columnas": ["disparos_totales", "disparos_en_area", "pases_ultimo_tercio_exitosos"],
       "pesos": [0.3, 0.4, 0.3]
   },
   "Solidez_Defensiva": {
       "columnas": ["goles_recibidos", "paradas_atrapando", "paradas_de_puños"],
       "pesos": [-0.5, -0.3, -0.2],  # Negativo para goles_recibidos porque menos es mejor
       "inverse": True
   },
   "Eficacia_Defensiva": {
       "columnas": ["intercepciones", "entradas_exitosas", "despejes", "bloqueos"],
       "pesos": [0.3, 0.3, 0.2, 0.2]
   },
   "Zonas_Recuperacion": {
       "columnas": ["recuperaciones_zona_alta", "recuperaciones_zona_media", "recuperaciones_zona_baja"],
       "pesos": [0.7, 0.25, 0.05]
   },
   "Presion_Alta": {
       "columnas": ["recuperaciones_zona_alta", "duelos_suelo_ganados_zona_alta", "entradas_ganadas_zona_alta"],
       "pesos": [0.4, 0.3, 0.3]
   }
}

def normalize_to_10(series, season_mean, season_std, inverse=False):
   z_scores = (series - season_mean) / season_std
   if inverse:
       return 5 - (z_scores * 2.5)
   return 5 + (z_scores * 2.5)

def calculate_kpis(df):
   kpi_df = pd.DataFrame()
   kpi_df['season_id'] = df['season_id']
   kpi_df['match_id'] = df['match_id']
   kpi_df['team_id'] = df['team_id']
   
   # Añadir la columna de equipo
   kpi_df['equipo'] = df['equipo']

   # Añadir la columna de partido
   kpi_df['partido'] = df['partido']
   
   # Añadir la columna de jornada
   kpi_df['jornada'] = df['jornada']

   # Calcular todos los KPIs
   for kpi_name, components in kpi_components.items():
       available_cols = [col for col in components['columnas'] if col in df.columns]
       if not available_cols:
           continue
           
       season_stats = df.groupby('season_id')[available_cols].agg(['mean', 'std'])
       kpi_values = pd.Series(0, index=df.index)
       
       for col, weight in zip(available_cols, components['pesos'][:len(available_cols)]):
           season_mean = df['season_id'].map(season_stats[col]['mean'])
           season_std = df['season_id'].map(season_stats[col]['std'])
           season_std = pd.Series([std if std != 0 else mean * 0.1 if mean != 0 else 1 
                                 for mean, std in zip(season_mean, season_std)], 
                                index=season_std.index)
           
           normalized_values = normalize_to_10(
               df[col], 
               season_mean, 
               season_std,
               inverse=components.get('inverse', False) and col == 'goles_recibidos'
           )
           kpi_values += normalized_values * abs(weight)
           
       kpi_df[kpi_name] = kpi_values.clip(0, 10)
   
   # Calcular KPI_Rendimiento como la media de todos los KPIs
   kpi_columns = [col for col in kpi_df.columns if col not in ['season_id', 'match_id', 'team_id', 'equipo', 'partido', 'jornada']]
   kpi_df['KPI_Rendimiento'] = kpi_df[kpi_columns].mean(axis=1)
   
   return kpi_df

# Crear el DataFrame de KPIs
df_KPI_equipos = calculate_kpis(df_equipos_estadisticas)

# Mostrar resultados
print("\nEstadísticas descriptivas de los KPIs:")
print(df_KPI_equipos.describe())

# Verificar valores nulos y fuera de rango
print("\nValores nulos por columna:")
print(df_KPI_equipos.isnull().sum())
```
